```python
import pandas as pd
pd.set_option('display.max_rows', 10000)
import numpy as np
import matplotlib.pyplot as plt
import os
import Levenshtein
```

```python
#df = pd.read_csv("../../ljspeech/unprocessed_sampled_asr.csv", sep=";")
df = pd.read_csv("../../ljspeech/unprocessed_sampled_original.csv")
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head(50)
```

```python
samples = df[df['Tag']=="LOC"]['Sentence #'].values.tolist()
```

```python
samples = list(set(samples))
```

```python
samples.sort()
```

```python
samples
```

```python
from itertools import groupby
a = [len(list(group)) for key, group in groupby(samples)]
```

```python
freq = pd.DataFrame()
freq['samples'] = samples
freq['frequency'] = a
```

```python
a
```

```python
freq.head()
```

```python
freq.sort_values('frequency', inplace=True, ascending=False)
```

```python
freq.head(100)
```

```python
new_samples = freq['samples'].values.tolist()
```

```python
new_samples
```

```python
sampled_df = df[df['Sentence #'].isin(new_samples[:100])]
```
